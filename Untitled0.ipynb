```python
import cv2
import easyocr
import matplotlib.pyplot as plt

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'], gpu=False)

# Set path to one image in the dataset (update this based on unzipped content)
image_path = '/content/textocr_data/train_val_images/train_images/c1309467624becbc.jpg'  # Corrected image path

# Read and convert image
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Extract text
results = reader.readtext(image_rgb)

# Print text results
for bbox, text, conf in results:
    print(f'"{text}" (Confidence: {conf:.2f})')

# Visualize results
for bbox, text, _ in results:
    p0 = tuple(map(int, bbox[0]))
    p2 = tuple(map(int, bbox[2]))
    cv2.rectangle(image_rgb, p0, p2, (0, 255, 0), 2)
    cv2.putText(image_rgb, text, p0, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("EasyOCR Result")
plt.show()
```

```python
import json, os, numpy as np
from PIL import Image

def load_annotations(json_path, images_dir):
    # Check if the file exists
    if not os.path.exists(json_path):
        print(f"Error: File not found at {json_path}")
        # List directory contents for debugging
        print(f"Contents of {os.path.dirname(json_path)}:")
        try:
            print(os.listdir(os.path.dirname(json_path)))
        except FileNotFoundError:
            print("Directory not found.")
        return [] # Return empty list if file not found

    data = json.load(open(json_path))
    imgs = {img['id']:img['file_name'] for img in data['images']}
    anns = {}
    for ann in data['annotations']:
        img_id = ann['image_id']
        text = ann['utf8_string']
        pts = np.array(ann['points'], dtype=int)
        xmin, ymin = pts[:,0].min(), pts[:,1].min()
        xmax, ymax = pts[:,0].max(), pts[:,1].max()
        bbox = [xmin, ymin, xmax, ymax]
        anns.setdefault(img_id, []).append((bbox, text))
    dataset = []
    for img_id, items in anns.items():
        img_path = os.path.join(images_dir, imgs[img_id])
        full_text = " ".join([t for _, t in items])
        dataset.append((img_path, full_text))
    return dataset

train_data = load_annotations('/content/textocr_data/train_val_images/train_images.jpg', '/content/textocr_data/train_val_images/train_images')
val_data   = load_annotations('/content/textocr_data/textocr/instances_val.json', '/content/textocr_data/train_val_images/train_images')

print(f"✅ Loaded {len(train_data)} training and {len(val_data)} validation samples.")
```

```python
from torch.utils.data import Dataset
from PIL import Image
import os
from tqdm import tqdm

class TextOCRDataset(Dataset):
    def __init__(self, annotations, image_dir, processor, max_samples=2000):
        self.samples = []
        self.processor = processor
        self.image_dir = image_dir

        print("🔍 Loading dataset samples...")
        for ann_id, ann in tqdm(annotations['anns'].items()):
            image_id = str(ann["image_id"])
            text = ann.get("utf8_string", "").strip()

            if not text:
                continue

            img_info = annotations['imgs'].get(image_id)
            if not img_info:
                continue

            file_name = img_info['file_name']  # e.g., "train_images/abc.jpg"
            image_path = os.path.join(image_dir, os.path.basename(file_name))

            if not os.path.exists(image_path):
                continue

            self.samples.append((image_path, text))

            if len(self.samples) >= max_samples:
                break

        print(f"✅ Loaded {len(self.samples)} samples")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        image_path, text = self.samples[idx]
        image = Image.open(image_path).convert("RGB")
        image = image.resize((224, 224)) # Resize image here
        encoded = self.processor(images=image, text=text, return_tensors="pt", padding="max_length", truncation=True)
        return {key: val.squeeze(0) for key, val in encoded.items()}
```

```python
from PIL import Image
import torch

def preprocess(example):
    # Open and resize image
    image = Image.open(example["file_path"]).convert("RGB")
    image = image.resize((224, 224))  # ✅ Required size for ViT

    # Process image to get pixel values
    pixel_values = processor(images=image, return_tensors="pt").pixel_values[0]

    # Tokenize the text label
    labels = processor.tokenizer(
        example["text"],
        padding="max_length",
        max_length=128,
        truncation=True,
        return_tensors="pt"
    ).input_ids[0]

    return {
        "pixel_values": pixel_values,
        "labels": labels
    }
```
